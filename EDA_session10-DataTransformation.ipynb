```python
import numpy as np
import matplotlib.pyplot as plt
#Read the data
exp_data=np.random.exponential(size=10000)
exp_data[:10]
```

```python
#histogram
plt.hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.show()
```

```python
norm_data=np.random.normal(size=1000)
plt.hist(norm_data)
```

```python
log_data=np.log(exp_data)
log_data[:10]
exp_data[:10]
plt.hist(log_data,bins=40,label='Logdata')
plt.legend()
plt.show()

```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=40,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=40,label='Log Transformation')
plt.legend()
plt.show()
```

```python
x=3
np.reciprocal(x)
rec_data=np.reciprocal(exp_data)
plt.hist(rec_data,bins=40,label='Reciprocal Data')
plt.legend()
plt.show()
exp_data,rec_data
```

```python
print(25**2) # square,
print(25**(1/2)) # square root
print(np.sqrt(25))
```

```python
sqrt_data=np.sqrt(exp_data)
sqrt_data
```

```python
plt.subplot(1,2,1).hist(exp_data,bins=40,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(sqrt_data,bins=40,label='Sqaure root Transformation')
plt.legend()
plt.show()
```

```python
import seaborn as sns
sns.displot(sqrt_data)
```

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
pt_data=pt.fit([exp_data]).transform([exp_data])
pt_data
plt.hist(pt_data[0])
plt.show()
```
