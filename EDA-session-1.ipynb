```python
#adding new column
name=['Ramesh','Suresh','Satish']
age=[30,35,40]
data=zip(name,age)
cols=['Name','Age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=cols,index=ind)
city_name=['Hyd','Blr','Chennai']
df['City']=city_name
df
```

```python
#update the existing column$
```

```python
df['Name']=['Lisi','Lipi','Liza']
df
```

```python
df['Age']=['25','35']
df
```

```python
#drop the column
df.drop('City',#column name
        axis=1,#column
        inplace=True)#overwrite coumn
df
```

```python
#drop the rows
df.drop('B',axis=0,inplace=True)
df
```

```python
#save the DataFrame
df.to_csv("output.csv")
```

```python
#read output.csv
pd.read_csv("output.csv")
```

```python
#To avoid the extra column in the output file we need to do index=False
df.to_csv("output.csv",index=False)
pd.read_csv("output.csv")
```
